```python
import pandas as pd

df = pd.read_csv('modified_hexaco_human.csv')

current_order = [6, 78, 30, 54, 53, 77, 29, 5, 4, 52, 28, 76, 51, 75, 3, 27, 50, 74, 2, 26, 1, 25, 49, 73, 
                 60, 12, 84, 36, 59, 11, 35, 83, 10, 34, 82, 58, 81, 33, 57, 9, 32, 56, 8, 80, 31, 79, 55, 7, 
                 66, 42, 18, 90, 17, 89, 41, 65, 64, 88, 40, 16, 39, 15, 87, 63, 14, 38, 62, 86, 61, 37, 85, 13, 
                 24, 72, 96, 48, 47, 95, 23, 71, 94, 70, 46, 22, 21, 93, 69, 45, 92, 68, 20, 44, 19, 67, 43, 91, 
                 97, 98, 99, 100]

column_mapping = {f'item{num}': f'item{i}' for i, num in enumerate(current_order)}

df.iloc[:, 4:104] = df.iloc[:, 4:104].rename(columns=column_mapping)

new_columns = [f'item{i}' for i in range(1, 101)]
df = pd.concat([df.iloc[:, :4], df[new_columns], df.iloc[:, 104:]], axis=1)

df.to_csv('reordered_hexaco_human.csv', index=False)

print(df.head())
```
